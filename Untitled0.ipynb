```python
from google.colab import files
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Загружаем изображение
uploaded = files.upload()
```

```python
# Получаем имя файла
filename = list(uploaded.keys())[0]
# Открываем изображение
img = Image.open(filename)
```

```python
# Показываем само изображение
plt.figure(figsize=(6, 6))
plt.imshow(img, cmap='gray')
plt.axis('on')
plt.show()
```

```python
# Переводим изображение в grayscale
gray = img.convert('L')
gray_np = np.array(img)
```

```python
# Строим гистограмму
plt.figure(figsize=(8, 4))
plt.hist(gray_np.flatten(), bins=256, range=(0, 255))
plt.title("Гистограмма яркости")
plt.xlabel("Уровень яркости")
plt.ylabel("Количество пикселей")
plt.show()
```

```python
# Создаём фигуру с двумя областями
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# 1) Изображение
ax1.imshow(gray, cmap='gray')
ax1.set_title("Изображение")
ax1.axis('off')

# 2) Гистограмма
ax2.hist(gray_np.flatten(), bins=256, range=(0, 255))
ax2.set_title("Гистограмма яркости")
ax2.set_xlabel("Яркость")
ax2.set_ylabel("Количество пикселей")

plt.tight_layout()
plt.show()
```
