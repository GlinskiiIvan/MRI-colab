```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Загрузка изображения
uploaded = files.upload()
filename = list(uploaded.keys())[0]
```

```python
# Чтение изображения в градациях серого
img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

# Проверка
if img is None:
    raise ValueError("Не удалось загрузить изображение!")

# --- Применение CLAHE ---
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
img_clahe = clahe.apply(img)
```

```python
# --- Построение графиков ---
plt.figure(figsize=(14, 8))

# Оригинал
plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Оригинал (Grayscale)")
plt.axis('off')

# CLAHE
plt.subplot(2, 2, 2)
plt.imshow(img_clahe, cmap='gray')
plt.title("CLAHE")
plt.axis('off')

# Гистограмма оригинала
plt.subplot(2, 2, 3)
plt.hist(img.ravel(), bins=256, range=(0, 256))
plt.title("Гистограмма оригинала")

# Гистограмма CLAHE
plt.subplot(2, 2, 4)
plt.hist(img_clahe.ravel(), bins=256, range=(0, 256))
plt.title("Гистограмма CLAHE")

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Загрузка изображения
uploaded = files.upload()
filename = list(uploaded.keys())[0]
```

```python
# Загружаем изображение (8-бит, grayscale)
img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

# Создаём сетку 2x4 для 8 битовых плоскостей
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
fig.suptitle("Битовые плоскости изображения", fontsize=16)

for k in range(1, 9):  # биты 1-8
    # Выделяем k-ю битовую плоскость
    bit_k = (img >> (k-1)) & 1
    # Масштабируем для отображения
    bit_k_vis = bit_k * 255

    # Определяем позицию в сетке
    ax = axes[(k-1)//4, (k-1)%4]
    ax.imshow(bit_k_vis, cmap='gray')
    ax.set_title(f"Бит {k}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```
